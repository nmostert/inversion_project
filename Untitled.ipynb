```python
def TGSD_SSE(tgsd, params, setup, z):
    tot_sum = 0
    contributions = []
    a, b, h1, wind_u, wind_v, \
        diffusion_coefficient, \
        fall_time_threshold, total_mass = params
    for phi_prob, stp in zip(tgsd, setup):
        m, n, p, z, z_min, elev, ft, \
            eddy_constant, samp_df, H, \
            fall_time_adj = stp
        
#         phi_prob = param_transform(phi_trans)
        
        phi_mass = total_mass*phi_prob

        A = get_plume_matrix(
            wind_u, wind_v, n, p, z, z_min, 
            elev, ft, diffusion_coefficient, fall_time_threshold, 
            eddy_constant, samp_df, H, fall_time_adj
            )
        beta_sum, beta_contributions= beta_sse([a, b, h1], A, z, m, phi_mass, z_min, H)
        contributions += [beta_contributions]        
        tot_sum += beta_sum
    print(tot_sum)
    return tot_sum, contributions

def TGSD_inversion(
    samp_df, num_samples, column_steps, 
    z_min, z_max, elevation, 
    phi_steps, params, eddy_constant=.04, tgsd_priors=None, 
    out="verb", column_cap=45000
):
    global AIR_VISCOSITY, GRAVITY, AIR_DENSITY

    layer_thickness = ((z_max-z_min)/column_steps)
    z = np.linspace(z_min + layer_thickness, z_max, column_steps)

    height_above_vent = z - z_min
    # TODO: This should be generalized to take point elevation into
    # account
    distance_below_vent = z_min - elevation

    # plume_diffusion_fine_particle = [column_spread_fine(ht) for ht in height_above_vent]
    # plume_diffusion_coarse_particle = [column_spread_coarse(ht, diffusion_coefficient) for ht in height_above_vent]


    setup = []
    coef_matrices = []
    
    for phi_step in phi_steps:
        d = phi2d(phi_step["lower"])/1000

        if distance_below_vent > 0:
            fall_time_adj = part_fall_time(
                    z_min, distance_below_vent, 
                    d, phi_step["density"],
                    AIR_DENSITY, 
                    GRAVITY, 
                    AIR_VISCOSITY
                )[0]
        else:
            fall_time_adj = 0.

        fall_values = [part_fall_time(
            zk, 
            layer_thickness,                   
            d, phi_step["density"], 
            AIR_DENSITY, 
            GRAVITY, 
            AIR_VISCOSITY
        ) for zk in z]


        terminal_velocities = [-e[1] for e in fall_values]
        fall_times = [e[0] for e in fall_values]

        m = samp_df["MassArea"].values \
            * (samp_df[phi_step["interval"]].values / 100)

        setup.append([
            m, num_samples, 
            column_steps, z, z_min, elevation, fall_times,
            eddy_constant, samp_df, column_cap,
            fall_time_adj
        ])
    
    # These are meant to be default (uninformed) guesses
    if tgsd_priors is None:
        tgsd_priors = [1/len(phi_steps)]*len(phi_steps)

    
    
#     tgsd_trans = [param_inv_transform(gsd) for gsd in tgsd_priors]
    
    prob_constraint = LinearConstraint([[1]*len(tgsd_priors)], [1], [1])
    
    bounds = Bounds([0]*len(tgsd_priors), [np.inf]*len(tgsd_priors))
    
    k0 = np.array(tgsd_priors, dtype=np.float64)

    def func(k):
        sse, cont = TGSD_SSE(k, params, setup, z)
        return sse
    
    # IT HAPPENS HERE
    sol = minimize(func, k0, method='trust-constr',  jac="2-point", hess=SR1(),
                   constraints=[prob_constraint], options={'verbose':1},
                   bounds=bounds)
    # THIS IS THE THING

#     params_out = [param_transform(p) for p in sol.x]
    
    sse, cont = TGSD_SSE(sol.x, params, setup, z)
    
    return sol.x, sol, sse, cont, setup, z

```

```python
params = [
    config["ALPHA"],
    config["BETA"],
    config["PLUME_HEIGHT"],
    u, v,
    config["DIFFUSION_COEFFICIENT"],
    config["FALL_TIME_THRESHOLD"],
    config["ERUPTION_MASS"]
]

data = obs_df

H = 45000
out, sol, sse_after, cont_after, setup, z = TGSD_inversion(
        data, len(data), 20, config["VENT_ELEVATION"], 
        H, 2500, phi_steps, params, tgsd_priors=tgsd,
        column_cap=H)
```
